```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense


# TODO 1: Load dataset
import pandas as pd
from sklearn.model_selection import train_test_split
from google.colab import files
uploaded = files.upload()
df = pd.read_csv(list(uploaded.keys())[0])
# Preview first 5 rows
print(df.head())

X = df[["HoursStudied","AssignmentsCompleted"]].values
y = df["ExamScore"].values

# TODO 2: Normalize X and y
X = X / np.max(X, axis=0)   # normalize features column-wise
y = y / np.max(y)           # normalize target

# TODO 3: Initialize Keras Sequential model
model = Sequential()

# TODO 4: Add input + hidden layers
model.add(Dense(5, input_dim=2, activation='relu'))
model.add(Dense(4, activation='relu'))

# TODO 5: Add output layer
model.add(Dense(1, activation='linear'))

# TODO 6: Compile model
model.compile(optimizer='adam', loss='mse')

# TODO 7: Train model
model.fit(X, y, epochs=50, batch_size=2, verbose=1)

# TODO 8: Evaluate model
loss = model.evaluate(X, y)
print("Keras Model Loss:", loss)

```

```python
import torch
import torch.nn as nn
import torch.optim as optim

# TODO 9: Convert to tensors
X_tensor = torch.tensor(X, dtype=torch.float32)
y_tensor = torch.tensor(y.reshape(-1,1), dtype=torch.float32)

# TODO 10: Define neural network
class SimpleNN(nn.Module):
    def __init__(self):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(2, 5)
        self.fc2 = nn.Linear(5, 4)
        self.fc3 = nn.Linear(4, 1)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        x = self.fc3(x)
        return x

# TODO 11: Initialize model, loss, optimizer
model_pt = SimpleNN()
criterion = nn.MSELoss()
optimizer = optim.Adam(model_pt.parameters(), lr=0.01)

# TODO 12: Train model
for epoch in range(50):
    # Forward pass
    outputs = model_pt(X_tensor)
    loss = criterion(outputs, y_tensor)

    # Backward pass
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item()}")

# TODO 13: Evaluate model
with torch.no_grad():
    predictions = model_pt(X_tensor)
    loss = criterion(predictions, y_tensor)
    print("PyTorch Model Loss:", loss.item())
```
