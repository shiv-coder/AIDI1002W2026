```python
# ============================================================
# WEEK 4  DATA VISUALIZATION SOLUTIONS
# ============================================================


# ============================================================
# Activity 1: Introduction to Visualization Libraries
# ============================================================

# TASK 1  Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# TASK 2  Create the DataFrame
data = {
    "Age": [18, 22, 25, 45, 52, 36, 24, 33, 40, 29],
    "Salary": [25000, 32000, 40000, 65000, 70000, 52000, 38000, 45000, 60000, 42000],
    "City": ["Toronto", "Toronto", "Ottawa", "Ottawa", "Toronto",
             "London", "London", "Ottawa", "London", "Toronto"]
}

df = pd.DataFrame(data)


# TASK 3  Explore Data
df.head()

df.info()


# ============================================================
# Activity 2: Matplotlib Basic Plots
# ============================================================

# A. Line Plot  Salary Trend
plt.plot(df["Salary"])
plt.title("Salary Trend")
plt.xlabel("Index")
plt.ylabel("Salary")
plt.show()


# B. Bar Chart  Average Salary by City
avg_salary = df.groupby("City")["Salary"].mean()

avg_salary.plot(kind="bar")
plt.title("Average Salary by City")
plt.xlabel("City")
plt.ylabel("Average Salary")
plt.show()


# C. Scatter Plot  Age vs Salary
plt.scatter(df["Age"], df["Salary"])
plt.title("Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()


# ============================================================
# Activity 3: Seaborn for Statistical Visuals
# ============================================================

# A. Histogram  Age Distribution
sns.histplot(df["Age"], kde=True)
plt.title("Age Distribution")
plt.show()


# B. Boxplot  Salary Spread
sns.boxplot(y=df["Salary"])
plt.title("Salary Spread")
plt.show()


# C. Pairplot  Multivariate Analysis
sns.pairplot(df)
plt.show()


# D. Heatmap  Correlation
corr = df[["Age", "Salary"]].corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()


# ============================================================
# Activity 4: Mini Visualization Project
# ============================================================

# Dataset
sales = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Units_Sold": [120, 150, 180, 160, 200, 220],
    "Profit": [4000, 5000, 5400, 4800, 6000, 6500]
}

sales_df = pd.DataFrame(sales)


# 1. Units Sold over Months  Line Chart
plt.plot(sales_df["Month"], sales_df["Units_Sold"])
plt.title("Units Sold Over Months")
plt.xlabel("Month")
plt.ylabel("Units Sold")
plt.show()


# 2. Profit over Months  Line Chart
plt.plot(sales_df["Month"], sales_df["Profit"])
plt.title("Profit Over Months")
plt.xlabel("Month")
plt.ylabel("Profit")
plt.show()


# 3. Bar Chart  Units vs Profit
plt.bar(sales_df["Month"], sales_df["Units_Sold"])
plt.title("Units Sold by Month")
plt.xlabel("Month")
plt.ylabel("Units Sold")
plt.show()


# 4. Scatter Plot  Units vs Profit
plt.scatter(sales_df["Units_Sold"], sales_df["Profit"])
plt.title("Units Sold vs Profit")
plt.xlabel("Units Sold")
plt.ylabel("Profit")
plt.show()


# 5. Combined Subplots
fig, axes = plt.subplots(2, 1, figsize=(8, 8))

axes[0].plot(sales_df["Month"], sales_df["Units_Sold"])
axes[0].set_title("Units Sold")

axes[1].plot(sales_df["Month"], sales_df["Profit"])
axes[1].set_title("Profit")

plt.tight_layout()
plt.show()


# STUDENT WRITING TASK  SAMPLE ANSWERS
# Which month had the highest units sold?
#  June
# What relationship do you see between profit and units sold?
#  Profit increases as units sold increase.
# Which chart was most helpful?
#  The line chart and scatter plot.


# ============================================================
# Activity 5: Chart Styling and Cleanup
# ============================================================

plt.figure(figsize=(8, 5))
sns.lineplot(x="Month", y="Profit", data=sales_df)
plt.title("Styled Profit Trend", fontsize=14)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Profit", fontsize=12)
plt.grid(True)
plt.show()


# STUDENT WRITING TASK  SAMPLE ANSWER
# I increased the figure size and added clear labels and a title.
# A grid was enabled to make values easier to read.
# Using Seaborn improved the visual appearance and clarity of the chart.
```
