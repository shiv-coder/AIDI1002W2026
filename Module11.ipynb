```python

#Step 1 — Load Dataset and Preview
import pandas as pd
from sklearn.model_selection import train_test_split

# TODO 1: Load dataset
from google.colab import files
uploaded = files.upload()
df = pd.read_csv(list(uploaded.keys())[0])
# Preview first 5 rows
print(df.head())
```

```python

# Define features (X) and target (y)
X = df[['HoursStudied', 'AssignmentsCompleted']].values
y = df['ExamScore'].values.reshape(-1, 1)
```

```python
#Part 2 — Activation Functions
# Sigmoid function
import numpy as np
def sigmoid(x):
    # TODO 3
    return 1 / (1 + np.exp(-x))

# ReLU function
def relu(x):
    # TODO 4
    return np.maximum(0, x)

# TODO 5: Test functions
sample = np.array([-2, -1, 0, 1, 2])

print("Sigmoid:", sigmoid(sample))
print("ReLU:", relu(sample))
```

```python
#Part 3 — Simple Neural Network (1 Hidden Layer)
# TODO 6: Initialize random weights
np.random.seed(0)

input_size = 2
hidden_size = 3
output_size = 1

W1 = np.random.rand(input_size, hidden_size)
b1 = np.random.rand(1, hidden_size)

W2 = np.random.rand(hidden_size, output_size)
b2 = np.random.rand(1, output_size)

# TODO 7 & 8: Forward propagation
# Hidden layer
z1 = np.dot(X, W1) + b1
a1 = relu(z1)

# Output layer
z2 = np.dot(a1, W2) + b2
output = sigmoid(z2)

# TODO 9: Compare predicted vs actual
print("Predicted Output:\n", output[:5])
print("Actual Output:\n", y[:5])
```

```python
#Part 4 — Functional Programming
# TODO 10: Add 5 bonus points to ExamScore
bonus_scores = list(map(lambda x: x + 5, df['ExamScore']))
print("Bonus Scores:", bonus_scores[:5])

# TODO 11: Filter students with ExamScore >= 75
high_scores = list(filter(lambda x: x >= 75, df['ExamScore']))
print("High Scores:", high_scores[:5])

# TODO 12: Square HoursStudied
squared_hours = list(map(lambda x: x**2, df['HoursStudied']))
print("Squared Hours:", squared_hours[:5])
```

```python
#Part 5 — Combine Neural Network with Functional Programming
# TODO 13: Normalize features using functional programming

# Normalize HoursStudied (divide by max)
max_hours = max(df['HoursStudied'])
norm_hours = list(map(lambda x: x / max_hours, df['HoursStudied']))

# Normalize AssignmentsCompleted
max_assign = max(df['AssignmentsCompleted'])
norm_assign = list(map(lambda x: x / max_assign, df['AssignmentsCompleted']))

# Combine normalized features
X_norm = np.array(list(zip(norm_hours, norm_assign)))

# Run through neural network again
z1 = np.dot(X_norm, W1) + b1
a1 = relu(z1)

z2 = np.dot(a1, W2) + b2
output_norm = sigmoid(z2)

print("Output after normalization:\n", output_norm[:5])
```
